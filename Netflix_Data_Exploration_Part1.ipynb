Replacing null value with the United States in country column
---



```python
nfdf['country'].replace(np.nan, 'United States', inplace=True)
nfdf.isnull().sum()
```

similarly replacing the null values in each column

```python
nfdf['director'].replace(np.nan, 'No Director', inplace=True)
nfdf['cast'].replace(np.nan, 'No Cast', inplace=True)
```

```python
nfdf['date_added'] = pd.to_datetime(nfdf['date_added'])
most_common_date = nfdf['date_added'].mode()[0]
nfdf['date_added'].fillna(most_common_date, inplace=True)
```

```python
nfdf.isnull().sum()
```

```python
nfdf['rating'].fillna('Not Rated', inplace=True)
```

```python
# Extract numeric part from 'duration' column and convert to float
nfdf['duration'] = nfdf['duration'].str.extract('(\d+)', expand=False).astype(float)

# Impute missing values for 'duration' with median
median_duration = nfdf['duration'].median()
nfdf['duration'].fillna(median_duration, inplace=True)
```


Individual Genres Trend

```python
genres = nfdf.set_index('title').listed_in.str.split(', ', expand=True).stack().reset_index(level=1, drop=True)
genres_df = pd.DataFrame()
genres_df['genre'] = genres
```

```python
# Reset index of years and description DataFrames
years = nfdf.set_index('title')['date_added'].reset_index()
description = nfdf.set_index('title')['description'].reset_index()

# Merge genres_df, years, and description DataFrames
temp = pd.merge(genres_df, years, on='title')
temp = pd.merge(temp, description, on='title')

# Plotting
plt.figure(figsize=(10, 10))
sns.countplot(y='genre', data=temp, order=temp['genre'].value_counts().iloc[:20].index)
plt.title('Top 10 Genres added by Netflix (2008 - 2021)')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.grid()
plt.show()
```

 Analyzing Top 5 TV Shows by Number of Seasons

```python
# Select relevant features
features = ['title', 'duration', 'type']
seasons = nfdf[features]

# Replace ' Season' and 's' with empty string to extract only the number of seasons
seasons['no_of_seasons'] = seasons['duration'].astype(str).str.replace(' Season', '').str.replace('s', '')

# Filter rows where 'type' is 'TV Show'
seasons = seasons[seasons['type'] == 'TV Show']

# Convert 'no_of_seasons' to numeric, handle errors by setting them to NaN
seasons['no_of_seasons'] = pd.to_numeric(seasons['no_of_seasons'], errors='coerce')

# Drop rows with NaN values in 'no_of_seasons'
seasons.dropna(subset=['no_of_seasons'], inplace=True)

# Convert 'no_of_seasons' to integer
seasons['no_of_seasons'] = seasons['no_of_seasons'].astype(int)

# Drop the 'duration' column
seasons = seasons.drop('duration', axis=1)

# Sort by number of seasons in descending order and select top 5
seasons = seasons.sort_values('no_of_seasons', ascending=False).head(5)

# Display the result
seasons
```
